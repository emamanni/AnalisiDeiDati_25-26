# Python for Data Science hands-on review: la libreria matplotlib

Matplotlib è una popolare libreria Python utilizzata per creare visualizzazioni statiche, animate e interattive. Fornisce un'ampia gamma di funzioni di plot per creare grafici, diagrammi e figure di alta qualità per l'analisi, l'esplorazione e la presentazione dei dati.

Plottare la funzione $f(x) = e^{-3x} + x$ ...

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 1000)
y = np.exp(-3*x) + x
plt.plot(x, y, 'red')
plt.show()
```

Aggiungere un titolo, etichette e annotazioni tra i comandi plt.plot e plt.show ...

```python
plt.plot(x, y, 'red')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axis([0, 5, 0, 5])
plt.grid(True)
plt.title('Funzione $e^{-3x} + x$')
plt.annotate('f(0)=1', xy=(0.005, 1), xytext=(0.2, 1.5), arrowprops=dict(arrowstyle='->'))
```

Plottare più funzioni ...

```python
y_2 =  2*x**2
plt.plot(x,y, 'b-', label='$f(x)=e^{-3x} + x$')
plt.plot(x,y_2, 'r--', label='$g(x)=2x^2$')
plt.legend(loc='lower right')
plt.xlabel('x')
plt.ylabel('f(x) e g(x)')
plt.axis([0, 5, 0, 5])
plt.grid(True)
plt.show()
```

Creazione di un istogramma ...

```python
x = [300, 500, 450, 800, 760, 550, 450, 470, 500, 650]
plt.hist(x, density=True, edgecolor='black', bins=10)
plt.xlabel('Dati')
plt.ylabel('Densità')
plt.show()
```

Creazione di un grafico a barre ...

```python
countries = ('Cina', 'India', 'USA', 'Indonesia', 'Pakistan', 'Italia')
population = [1439, 1380, 331, 273, 220, 60]
x_pos = np.arange(6)
plt.bar(x_pos, population, color='pink', edgecolor='black', align='center')
plt.xticks(x_pos, countries)
plt.ylabel('Popolazione (milioni)')
plt.title('Nazioni del mondo per popolazione (2021)')
plt.show()
```

Creazione di un grafico a barre orizzontale ...

```python
plt.barh(x_pos, population, color='pink', edgecolor='black', align='center')
plt.yticks(x_pos, countries)
plt.xlabel('Popolazione (milioni)')
plt.title('Nazioni del mondo per popolazione (2021)')
plt.show()
```

Creazione di un boxplot ...

```python
x = [300, 500, 450, 800, 760, 550, 450, 470, 500, 650, 1500]
plt.boxplot(x, showmeans=True)
plt.show()
```
